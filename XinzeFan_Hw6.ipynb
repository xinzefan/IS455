```python
# homework coorprate with ZhengYang Xu and JinPei Zou
# we work togethr through zoom and communicated about some methods
import pandas as pd
import bqplot
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
```

```python
# reading csv
buildings = pd.read_csv("building_inventory.csv")
```

```python
# select the new dataframe with drop the unuseful Year Acquired
nb = buildings[buildings['Year Acquired']!= 0]
nb
```

```python
# only select the last version's information
nb = nb.drop_duplicates('Location Name',keep = 'last')
# create new dataframe after groupby which given the sum of the square footage
data = nb[['Agency Name', 'Congress Dist', 'Square Footage']]
data = data.groupby(['Agency Name','Congress Dist'])['Square Footage'].sum().reset_index(name='Total Square Footage')
lst = data['Agency Name'].unique().tolist()
data.sort_values(by='Agency Name')
def set_index(x):
    if x in lst:
        return lst.index(x)
```

```python
# created the new sorted data with the distinct congress dist and agency name
row = sorted(data['Congress Dist'].unique().tolist())
col = data['Agency Name'].unique().tolist()
new = pd.DataFrame(columns= col, index = row )
# put the footage into the match congress dist, agency name
for index in data.index:
    new.loc[data['Congress Dist'][index],data['Agency Name'][index]] = data['Total Square Footage'][index]
# the na part need to be fill with 0 in order to groupby later 
new = new.fillna(0)
new
```

```python
mySelectedLabel = ipywidgets.Label()
#### plot part same as the steps in class
x_sc = bqplot.OrdinalScale() # categorical data
y_sc = bqplot.OrdinalScale()
col_sc = bqplot.ColorScale(scheme='Reds')

# 3. Axes
ax_x = bqplot.Axis(scale=x_sc)
ax_y = bqplot.Axis(scale=y_sc, orientation='vertical')
ax_col = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right') # colorbar

# 4. Mark (gridheat map)
heat_map = bqplot.GridHeatMap(color = np.array(new),
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'}, # on clicking on a grid, select that grid
                             anchor_style={'fill':'blue'}, # turn the grid blue on selection
                             selected_style={'opacity':1.0}, # selected grid is fully opaque on selection,
                             unselected_style={'opacity':0.8}) # otherwise, a little see-through

```

```python
# initial the original line plot
i,j = 1,1
col[j]
n1 = nb[nb['Agency Name'] == col[j]]
lind = n1[n1['Congress Dist'] == i]
lind
```

```python
# line plot
x_sch = bqplot.LinearScale()
y_sch = bqplot.LinearScale()
x_axh = bqplot.Axis(scale=x_sch,label = "Year")
y_axh = bqplot.Axis(scale=y_sch, label='Total Footage', orientation='vertical')
#sub = nb[nb['Congress Dist']== i & nb['Agency Name'] == col[j]]
#calculated each years sum square footage
sn = lind.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Footage')

line = bqplot.Lines(x=sn['Year Acquired'], y=sn['Total Footage'] , scales={'x': x_sch, 'y': y_sch},
              colors=['red'])
figh = bqplot.Figure(marks=[line], axes=[x_axh, y_axh])
```

```python
# selected action part
def on_selected(change):
    if len(change['owner'].selected) == 1: # forcing only selection of 1 grid
        i,j = change['owner'].selected[0] # only 1 selection
        v = new.iloc[i,j] # value of mean of the data at the i,j grid
        mySelectedLabel.value = 'Total Square Footage =' + str(v)
        # replot the line plot
        lind =nb[(nb['Agency Name'] == col[j]) & (nb['Congress Dist'] == i)]
        sn = lind.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Footage')
        line.x = sn['Year Acquired']
        line.y = sn['Total Footage']

heat_map.observe(on_selected,'selected')

# 5. Interactions (also in step 4 as well)
fig = bqplot.Figure(marks=[heat_map], axes=[ax_col, ax_x, ax_y])
fig.layout.min_width='800px'
figh.layout.min_width='800px'
figures = ipywidgets.HBox([fig,figh])
myDashboard = ipywidgets.VBox([mySelectedLabel, figures])

myDashboard
```
